```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most? 
#5 & Manhattan shows up 21,661 times
neighbourhood_groups_count = air_bnb.groupby(['neighbourhood_group']).count()
neighbourhood_groups_count
```

```python
# Are private rooms the most popular in manhattan?
#No, Entire home/apt is with 25,409 (vs 22,326 for Private room)
room_type_count = air_bnb.groupby(['room_type']).count()
room_type_count
```

```python
# Which hosts are the busiest and based on their reviews? 
# Host id: 47621202 (Dona) has the most reviews, but availability is 253 on avg.
# The busiest host is host id:107434423 (Blueground), but only has 29 reviews, working 25796 listing days across  232 host listings
host_id_mean_reviews = air_bnb.groupby(['host_id']).mean().sort_values(by = 'number_of_reviews', ascending = False)
host_id_mean_reviews

air_bnb['busy_days'] = 365 - air_bnb['availability_365']
air_bnb
host_id_busy_days = air_bnb.groupby(['host_id']).sum().sort_values(by = 'busy_days', ascending = False)
host_id_busy_days
host_id_busy_days_count = air_bnb.groupby(['host_id']).count().sort_values(by = 'busy_days', ascending = False)
host_id_busy_days_count
air_bnb.loc[air_bnb['host_id'] == 47621202]
air_bnb.loc[air_bnb['host_id'] == 107434423]
```

```python
#Which neighorhood group has the highest average price? 
#Manhattan at $196.88
neighbourhood_groups_mean_price = air_bnb.groupby(['neighbourhood_group']).mean().sort_values(by = 'price', ascending = False)
neighbourhood_groups_mean_price
```

```python
# Which neighbor hood group has the highest total price? 
#Manhattan, the total price would be: $4,264,527
neighbourhood_groups_sum_price = air_bnb.groupby(['neighbourhood_group']).sum().sort_values(by = 'price', ascending = False)
neighbourhood_groups_sum_price
```

```python
#Which top 5 hosts have the highest total price?
'''Host Id's: 
219517861 with $82,795 - Sonder (NYC)
107434423 with $70,331 - Blueground
156158778 with $37,097 - Sally
205031545 with $35,294 - Red Awning
30283594  with $33,581 - Kara
'''
host_id_sum_price = air_bnb.groupby(['host_id']).sum().sort_values(by = 'price', ascending = False)
host_id_sum_price
air_bnb.loc[air_bnb['host_id'] == 219517861]
air_bnb.loc[air_bnb['host_id'] == 107434423]
air_bnb.loc[air_bnb['host_id'] == 156158778]
air_bnb.loc[air_bnb['host_id'] == 205031545]
air_bnb.loc[air_bnb['host_id'] == 30283594]
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
# 162 hosts, see group returned for specifics
Review100_avail0 = air_bnb[(air_bnb['availability_365'] == 0) & (air_bnb['number_of_reviews'] >= 100)]
Review100_avail0
Review100_avail0_grouped = Review100_avail0.groupby('host_id').filter(lambda group: (group['availability_365'] == 0).all() and (group['number_of_reviews'] >= 100).all())
Review100_avail0_grouped
```

```python
# What host has the highest total of prices and where are they located?
# 219517861 with $82,795 - Sonder (NYC) (from earlier question)
# They are located in Manhattan 
air_bnb.loc[air_bnb['host_id'] == 219517861]

```
